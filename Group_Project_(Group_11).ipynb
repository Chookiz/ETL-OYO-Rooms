```python
def scrape_product_page(url):
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
  response = requests.get(url, headers=headers)
  data = response.text
  hotel_no = 0
  hotels_list = {}
  if response.status_code == 200:
        soup = BeautifulSoup(data, "html.parser")
        review = soup.find("p",{"class":"c-1wawhq6"}).text
        title = soup.find("h1",{"class":"c-1wj1luj"}).text
        location = soup.find("span",{"itemprop":"streetAddress"}).text
        defprice = soup.find("span",{"class":"listingPrice__slashedPrice d-body-lg"}).text
        discprice = soup.find("span",{"class":"listingPrice__finalPrice listingPrice__finalPrice--black"}).text
        rating = soup.find("span",{"class":"c-1uxth7l"}).text
        room = soup.find("span",{"class":"c-2j9z2q"}).text
        amenities = soup.find_all("div",{"class":"c-12w6zty"})
        combined_amenities = []

        for amenity in amenities:
          combined_amenities.append(amenity.text)

        final_amenities = ', '.join(combined_amenities)

        hotel_no += 1
        hotels_list[hotel_no] = [title,location,final_amenities,defprice,discprice,room,rating,review]
        hotels_df = pd.DataFrame.from_dict(hotels_list, orient ="index", columns=['Hotel Name','Location','Amenities','Default Price','Discounted Price','Room Type','Rating','Review'])
        return hotels_df
  else:
        print(f"Failed to fetch {url}. Status code: {response.status_code}")
```

```python
# Iterate through the URLs and call the scrape_product_page function
hotels_df=pd.DataFrame(columns=['Hotel Name','Location','Amenities','Default Price','Discounted Price','Room Type','Rating','Review'])
for url in urls:
  try:
        hotel_data = scrape_product_page(url)
        hotels_df = pd.concat([hotels_df, hotel_data], ignore_index=True)
  except Exception as e:
        print(f"Skipping {url} due to error: {e}")
hotels_df
```

```python
# Removing currency symbols (â‚¹), RM and Rp in column, 'default price' and 'discounted price'
df[['Default Price', 'Discounted Price']] = df[['Default Price', 'Discounted Price']].replace('[\â‚¹,]', '', regex=True)
df[['Default Price', 'Discounted Price']] = df[['Default Price', 'Discounted Price']].replace('RM', '', regex=True)
df[['Default Price', 'Discounted Price']] = df[['Default Price', 'Discounted Price']].replace('Rp', '', regex=True)
```

```python
#Remove one '.' if there are two.
# Define a condition for strings with two periods in 'Default Price'
condition_default = df['Default Price'].str.count('\.') == 2

# Apply the modification only to the cells meeting the condition for 'Default Price'
df.loc[condition_default, 'Default Price'] = df.loc[condition_default, 'Default Price'].apply(lambda x: x.replace('.', '', 1))

# Define a condition for strings with two periods in 'Discounted Price'
condition_discounted = df['Discounted Price'].str.count('\.') == 2

# Apply the modification only to the cells meeting the condition for 'Discounted Price'
df.loc[condition_discounted, 'Discounted Price'] = df.loc[condition_discounted, 'Discounted Price'].apply(lambda x: x.replace('.', '', 1))
```

```python
#remove patterns
import re
def remove_patterns(text):
    patterns = [
        r'OYO \d*',
        r'SPOT ON \d*',
        r'Super OYO Capital O \d*',
        r'Capital O \d*',
        r'Collection O \d*',
        r'Super'
    ]
    for pattern in patterns:
        text = re.sub(pattern, '', text)
    return text.strip()
```

```python
#remove useless pattern in hotel names
df['Hotel Name'] = df['Hotel Name'].apply(remove_patterns)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import string
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Tokenization function
def tokenize_text(text):
    # Tokenization
    tokens = word_tokenize(text)
    return tokens
```

```python
#Text Processing
def process_text(tokens):
    # Lowercasing
    tokens = [word.lower() for word in tokens]

    # Removing Punctuation
    tokens = [word for word in tokens if word.isalnum()]

    # Removing Stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    return tokens
```

```python
# Word Frequency Analysis
def plot_word_frequency(tokens):
    # Calculate word frequency
    freq_dist = FreqDist(tokens)

    # Convert frequency distribution to DataFrame
    freq_df = pd.DataFrame(list(freq_dist.items()), columns=['Word', 'Frequency'])

    # Sort by frequency in descending order
    freq_df = freq_df.sort_values(by='Frequency', ascending=False)

    # Plot word frequency using Seaborn
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Frequency', y='Word', data=freq_df.head(20), palette='viridis')
    plt.title('Words Frequency')
    plt.xlabel('Frequency')
    plt.ylabel('Word')
    plt.show()
```

```python
# Apply tokenization to the 'Review' column
df['Tokenized_Review'] = df['Review'].apply(tokenize_text)
```
